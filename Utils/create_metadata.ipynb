```python
import torchaudio
import os
import pandas as pd
import IPython.display as ipd
```

```python
VOXCELEB_PATH = 'VoxCelebTest'

paths = os.listdir(VOXCELEB_PATH)
paths2 = [ os.listdir(VOXCELEB_PATH + '/' + paths[i] ) for i in range(len(paths)) ]
paths3 = [ 
            [ 
                os.listdir(VOXCELEB_PATH + '/' + paths[i] + '/' + paths2[i][j])
                for j in range(len(paths2[i]))
            ]    
            for i in range(len(paths)) 
         ]

idx1, idx2, idx3 = 0, 0, 12
audio_file = VOXCELEB_PATH + '/' + paths[idx1] + '/' + paths2[idx1][idx2] + '/' + paths3[idx1][idx2][idx3]

waveform, sample_rate = torchaudio.load(audio_file)
ipd.Audio(waveform.numpy(), rate=sample_rate)
```

```python
df = pd.DataFrame(columns=['path', 'speaker_id', 'audio_id'])

for i in range(len(paths)):
    for j in range(len(paths2[i])):
        for k in range(len(paths3[i][j])):
            path = VOXCELEB_PATH + '/' + paths[i] + '/' + paths2[i][j] + '/' + paths3[i][j][k]
            df = pd.concat([df, pd.DataFrame([[path, paths[i], paths2[i][j]]], columns=['path', 'speaker_id', 'audio_id'])], ignore_index=True)

df.to_csv('metadata.csv', index=False)
```
